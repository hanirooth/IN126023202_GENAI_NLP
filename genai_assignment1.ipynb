```python
# =========================================
# 📌 NLP ASSIGNMENT - COMPLETE NOTEBOOK
# =========================================

# =========================================
# ✅ TASK 1: CONCEPTUAL UNDERSTANDING
# =========================================

print("TASK 1 ANSWERS:\n")

print("1. Difference between 'Love' and 'love':")
print("- NLP treats them as different unless lowercased.")
print("- After preprocessing → both become 'love'\n")

print("2. If stopwords are not removed:")
print("- Data becomes noisy")
print("- Model performance decreases\n")

print("3. Harmful stopword removal cases:")
print("- Sentiment: 'not happy' → removing 'not' changes meaning")
print("- Questions: 'What is time?' → becomes 'time'\n")

print("4. Stemming vs Lemmatization:")
print("- Stemming: cuts words (running → runn)")
print("- Lemmatization: meaningful root (running → run)\n")


# =========================================
# ✅ TASK 2: PREPROCESS FUNCTION
# =========================================

import re

def preprocess_text(text):
    try:
        if not text or text.strip() == "":
            return [], ""

        # Remove URLs & emails
        text = re.sub(r"http\S+|www\S+|@\S+", "", text)

        # Remove numbers
        text = re.sub(r"\d+", "", text)

        # Lowercase
        text = text.lower()

        # Handle repeated characters
        text = re.sub(r'(.)\1{2,}', r'\1\1', text)

        # Remove special characters
        text = re.sub(r"[^a-z\s]", "", text)

        # Remove extra spaces
        text = re.sub(r"\s+", " ", text).strip()

        tokens = text.split()

        # Remove short words except 'no', 'not'
        tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

        cleaned_sentence = " ".join(tokens)

        return tokens, cleaned_sentence

    except:
        return [], ""


# =========================================
# ✅ TASK 3: STRESS TESTING
# =========================================

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

print("\n\nTASK 3 OUTPUT:\n")

for text in test_sentences:
    tokens, cleaned = preprocess_text(text)
    results.append((text, tokens, cleaned))

    print("Original:", text)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-" * 50)


# =========================================
# ✅ TASK 4: TOKEN ANALYTICS
# =========================================

def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, round(avg_len, 2)


print("\n\nTASK 4 ANALYTICS:\n")

for text, tokens, _ in results:
    total, unique, avg_len = token_analytics(tokens)

    print("Sentence:", text)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", avg_len)
    print("-" * 50)


print("\nAnalysis Answers:")
print("Most noisy sentence: Win $$$ now!!! Limited offer!!!")
print("Most meaningful sentence: I absolutely looooved this product")


# =========================================
# ✅ TASK 5: FREQUENCY ANALYSIS
# =========================================

from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("\n\nTASK 5 FREQUENCY ANALYSIS:\n")

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[::-1][:5])


# =========================================
# ✅ TASK 6: FULL PIPELINE
# =========================================

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

pipeline_output = full_pipeline(test_sentences)

print("\n\nTASK 6 PIPELINE OUTPUT:\n")
print(pipeline_output)


# =========================================
# ✅ TASK 7: ERROR HANDLING
# =========================================

print("\n\nTASK 7 ERROR HANDLING:\n")

test_cases = ["", "😍😍😍", "123456"]

for t in test_cases:
    tokens, cleaned = preprocess_text(t)
    print("Input:", t)
    print("Tokens:", tokens)
    print("Cleaned:", cleaned)
    print("-" * 50)


# =========================================
# 🎯 END OF NOTEBOOK
# =========================================
```
